```python
import matplotlib.pyplot as plt
Partition = 'Holidays', 'Eating_Out', 'Shopping', 'Groceries'
sizes = [250, 100, 300, 200]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=Partition, autopct='%1.1f%%', shadow=True, startangle=90)         
ax1.axis('equal')
plt.show()
```

I think this python code is very interesting because you are able to draw a pie chart using python code. Instead of jus telling people the probability of the data you are trying to show you can use a graph like this to show people visually. It is easier to understand and breaks apart the different sections you are trying to compare inside the graph. With a quick look you can easily tell that shopping has the biggest partition here while eating out has the lowest.
